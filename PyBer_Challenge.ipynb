# Pyber Challenge

### 4.3 Loading and Reading CSV files

```python
# Add Matplotlib inline magic command
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd

# File to Load (Remember to check these)
city_data_to_load = "Resources/city_data.csv"
ride_data_to_load = "Resources/ride_data.csv"

# Read the City and Ride Data
city_data_df = pd.read_csv(city_data_to_load)
ride_data_df = pd.read_csv(ride_data_to_load)
```

### Merge the DataFrames

```python
# Combine the data into a single dataset
pyber_data_df = pd.merge(ride_data_df, city_data_df, how="left", on=["city", "city"])

# Display the data table for preview
pyber_data_df.head()
```

## Challenge Deliverable 1. Generate a Ride-Sharing DataFrame by City Type

```python
#  1. Get the total rides for each city type
sum_rides_per_city = pyber_data_df.groupby(["type"]).count()["ride_id"]
sum_rides_per_city
```

```python
# 2. Get the total drivers for each city type
total_drivers_by_city = city_data_df.groupby(["type"])["driver_count"].sum()
total_drivers_by_city 
```

```python
#  3. Get the total amount of fares for each city type
sum_fares_by_type = pyber_data_df.groupby(["type"])["fare"].sum()
sum_fares_by_type
```

```python
#  4. Get the average fare per ride for each city type. 
avg_fares_by_type = pyber_data_df.groupby(["type"])["fare"].mean()
avg_fares_by_type 
```

```python
total_fares_by_city = pyber_data_df.groupby(["type"])["fare"].sum()
total_fares_by_city
```

```python
total_rides_by_city = pyber_data_df.groupby(["type"])["ride_id"].count()
total_rides_by_city
```

```python
# 5. Get the average fare per driver for each city type.
#In Step 5, calculate the average fare per driver by city type by dividing the sum of all the fares by the total drivers.
total_fares_by_city = pyber_data_df.groupby(["type"])["fare"].sum()
avg_fare_per_driver_each_city = total_fares_by_city /total_drivers_by_city
avg_fare_per_driver_each_city
```

```python
#  6. Create a PyBer summary DataFrame. 
pyber_summary_df = pd.DataFrame({             
             "Total Rides": sum_rides_per_city,
             "Total Drivers": total_drivers_by_city,
             "Total Fares":sum_fares_by_type,
             "Average Fare per Ride":avg_fares_by_type,
             "Average Fare per Driver":avg_fare_per_driver_each_city})
pyber_summary_df 
```

```python
#  7. Cleaning up the DataFrame. Delete the index name
pyber_summary_df.index.name = None
```

```python
#  8. Format the columns.
pyber_summary_df ["Total Fares"] = pyber_summary_df["Total Fares"].map("${:,.2f}".format)
pyber_summary_df ["Average Fare per Ride"] = pyber_summary_df["Average Fare per Ride"].map("${:,.2f}".format)
pyber_summary_df ["Average Fare per Driver"] = pyber_summary_df["Average Fare per Driver"].map("${:,.2f}".format)
pyber_summary_df
```

## Deliverable 2.  Create a multiple line plot that shows the total weekly of the fares for each type of city.

```python
# Print the merged DataFrame for reference.
pyber_data_df.head()
```

```python
# 1. Using groupby() to create a new DataFrame showing the sum of the fares 
#  for each date where the indices are the city type and date.
new_pyber_data_df = pyber_data_df.copy()
total_fares_by_cities_and_dates = new_pyber_data_df.groupby(["type","date"])["fare"].sum()
total_fares_by_cities_and_dates_df = pd.DataFrame(total_fares_by_cities_and_dates)
total_fares_by_cities_and_dates_df
```

```python
# 2. Reset the index on the DataFrame you created in #1. This is needed to use the 'pivot()' function.
reset_index_df = total_fares_by_cities_and_dates_df.reset_index()
reset_index_df
```

```python
# 3. Create a pivot table with the 'date' as the index, the columns ='type', and values='fare' 
# to get the total fares for each type of city by the date.
pivot_table_df = pd.pivot_table(reset_index_df, index=["date"], values="fare", columns = ["type"])
pivot_table_df.head(10)
```

```python
# 4. Create a new DataFrame from the pivot table DataFrame using loc on the given dates, '2019-01-01':'2019-04-28'.
new_pivot_dataframe = pivot_table_df.loc['2019-01-01':'2019-04-28']
new_pivot_dataframe.head(10)
```

```python
# 5. Set the "date" index to datetime datatype. This is necessary to use the resample() method in Step 8.
new_pivot_dataframe.index = pd.to_datetime(new_pivot_dataframe.index)
new_pivot_dataframe
```

```python
# 6. Check that the datatype for the index is datetime using df.info()
new_pivot_dataframe.info()
```

```python
# 7. Create a new DataFrame using the "resample()" function by week 'W' and get the sum of the fares for each week.
new_pivot_dataframe_week = new_pivot_dataframe.resample('w').sum()
new_pivot_dataframe_week.head(10)
```

```python
# 8. Using the object-oriented interface method, plot the resample DataFrame using the df.plot() function. 

# Add all ride count box-and-whisker plots for the urban, suburban, rural fare data.
from matplotlib import style
style.use('fivethirtyeight')
# Create the plot and add a label for the legend.
ax = new_pivot_dataframe_week.plot(figsize=(15,5))
ax.set_title('Total Fare by City Type', fontsize=17)
ax.set_ylabel('Fare($USD)',fontsize=12)
ax.set_xlabel(' ')
ax.grid(True)
lgnd = ax.legend()
# Create a legend
lgnd = ax.legend(fontsize="12", mode="Expanded",
         loc="center", title="type")
lgnd.legendHandles[0]._sizes = [75]
lgnd.legendHandles[1]._sizes = [75]
lgnd.legendHandles[2]._sizes = [75]
lgnd.get_title().set_fontsize(12)
plt.savefig("analysis/PyBer_fare_summary.png")
# Show the plot
plt.show()
```
